```python
import argparse
import gym

def build_arg_parser():
    parser = argparse.ArgumentParser(description='Run an environment')
    parser.add_argument(
        '--input-env',
        dest='input_env',
        default='cartpole',  # Valeur par d√©faut pour √©viter l'erreur dans Jupyter
        choices=['cartpole', 'mountaincar', 'pendulum'], 
        help='Specify the name of the environment'
    )
    return parser

def get_args():
    if "__file__" not in globals():
        class Args:
            input_env = "cartpole"  # Remplacez par l‚Äôenvironnement voulu
        return Args()
    else:
        return build_arg_parser().parse_args()

if __name__ == '__main__':
    args = get_args()
    input_env = args.input_env

    name_map = {
        'cartpole': 'CartPole-v1',  # Gym a mis √† jour CartPole-v0 vers v1
        'mountaincar': 'MountainCar-v0',
        'pendulum': 'Pendulum-v1'
    }

    # Cr√©er l‚Äôenvironnement avec mode de rendu "human"
    env = gym.make(name_map[input_env], render_mode="human")

    print(f"üöÄ Ex√©cution de l'environnement : {input_env}")

    for episode in range(20):
        # R√©initialiser l‚Äôenvironnement
        observation, _ = env.reset()  # Dans Gym v26+, reset() renvoie (observation, info)
        
        for step in range(100):
            # Afficher l'environnement (attention : ne fonctionne pas dans Jupyter)
            env.render()

            # Afficher l'√©tat actuel
            print(f"√âpisode {episode+1}, √âtape {step+1}, Observation : {observation}")

            # Prendre une action al√©atoire
            action = env.action_space.sample()

            # Effectuer l'action et obtenir la nouvelle observation
            observation, reward, done, _, _ = env.step(action)

            # V√©rifier si l'√©pisode est termin√©
            if done:
                print(f"‚úÖ √âpisode termin√© apr√®s {step+1} √©tapes\n")
                break

    env.close()  # Fermer proprement
    print("üîÑ Simulation termin√©e.")

```
